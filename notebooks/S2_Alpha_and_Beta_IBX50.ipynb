```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ibov = pd.read_csv('../data/raw/^IBX50.csv', index_col=0, parse_dates=True)
ibov.head()
```

```python
ibov['^IBX50'] = ibov['Close'].pct_change(1)
```

```python
index = pd.read_csv('../data/interim/ibovfull.csv', index_col=0, parse_dates=True)
```

```python
itub = index[index['Name'] == 'ITUB4'].copy()
```

```python
itub['ITUB4'] = itub['Close'].pct_change(1)
```

```python
joined = itub[['ITUB4']].join(ibov['^IBX50'])
```

```python
joined.head()
```

```python
joined.iloc[100:150].plot(figsize=(10, 5));
```

```python
joined.plot.scatter('^IBX50', 'ITUB4')
```

```python
# make the dataset
joined.dropna(inplace=True)
X = joined[['^IBX50']].to_numpy()
Y = joined[['ITUB4']].to_numpy()
```

```python
X.shape, Y.shape
```

```python
plt.scatter(X, Y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
```

```python
X_predict = np.linspace(X.min(), X.max(), 5).reshape(-1, 1)
Y_predict = model.predict(X_predict)

plt.scatter(X, Y);
plt.xlabel('^IBX50');
plt.ylabel('ITUB4');
plt.plot(X_predict, Y_predict);
```

```python
beta = model.coef_
alpha = model.intercept_
alpha, beta
```

```python
# Manual calulation

Y_predict = beta * X_predict + alpha

plt.scatter(X, Y);
plt.xlabel('^IBX50');
plt.ylabel('ITUB4');
plt.plot(X_predict, Y_predict);
```
