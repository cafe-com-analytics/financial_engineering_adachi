Exercise 1
1.Using numpy, generate 1000 values from the normal random distribution ($\mu=0$, $\sigma=1$).
2.Plot the time series.
3.Plot the histogram.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
ts = np.random.standard_normal(1000)
ts = pd.DataFrame(ts, columns=['random'])
ts.head()
```

```python
plt.plot(ts)
```

```python
n, bins, patches = plt.hist(x=ts, bins='auto', alpha=0.7, rwidth=0.85)

plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exercise 1')
```

Exercise 2
    1. Add a trend line to the noise.
    2. Make a scatterplot.

```python
ts['x'] = np.linspace(0, 10, 1000)
```

```python
ts['y'] = ts['x'] + ts['random']
```

```python
ts.head()
```

```python
fit = np.polyfit(ts['x'],ts['y'],1)
ang_coeff = fit[0]
intercept = fit[1]
fit_eq = ang_coeff*ts['x'] + intercept
```

```python
#Plotting the data
fig = plt.figure()
ax = fig.subplots()
ax.plot(ts['x'], fit_eq,color = 'r', alpha = 0.5, label = 'Linear fit')
ax.scatter(ts['x'],ts['y'],s = 5, color = 'b', label = 'Data points') #Original data points
ax.set_title('Linear fit example')
ax.legend()
plt.show()
```

Exercise 4
1. Generate and plot 1000 samples from the multivariate normal
2. use Scipy or any onther library you prefer

```python
from scipy.stats import multivariate_normal
```

```python
mean = [0, 0]
cov = [[1, -0.5], [-0.5, 2]]  # diagonal covariance
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, '.')
plt.show()
```

```python
def cov(a, b):

    if len(a) != len(b):
        return

    a_mean = np.mean(a)
    b_mean = np.mean(b)

    sum = 0

    for i in range(0, len(a)):
        sum += ((a[i] - a_mean) * (b[i] - b_mean))

    return sum/(len(a)-1)
```

```python
cov(x,y)
```
